### Given a set of blue points, calculate the parameters of the fitted quadratic curve using the gradient descent method.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 定义二次曲线模型
def quadratic_model(x, theta):
    return theta[0] + theta[1] * x + theta[2] * x**2

# 计算损失函数
def compute_loss(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

# 计算梯度
def compute_gradient(x, y, theta):
    n = len(x)
    y_pred = quadratic_model(x, theta)
    grad = np.zeros(3)
    grad[0] = -2 * np.sum(y - y_pred)
    grad[1] = -2 * np.sum((y - y_pred) * x)
    grad[2] = -2 * np.sum((y - y_pred) * x**2)
    return grad

# 梯度下降算法
def gradient_descent(x, y, learning_rate=0.002, epochs=200):
    theta = np.random.randn(3)  # 初始化参数
    losses = []
    for epoch in range(epochs):
        grad = compute_gradient(x, y, theta)
        theta -= learning_rate * grad
        y_pred = quadratic_model(x, theta)
        loss = compute_loss(y, y_pred)
        losses.append(loss)
        if epoch % 100 == 0:
            print(f"Epoch {epoch}, Loss: {loss}")
    return theta, losses
```

```python
np.random.seed(42)

# 生成数据
x = np.linspace(-1, 1, 100)
y = 1.5 * x**2 + 0.2 + 0.2 * np.random.randn(100)

# 进行梯度下降拟合
theta, losses = gradient_descent(x, y, learning_rate=0.0001, epochs=2000)

# 打印拟合的参数
print("Fitted Parameters:", theta)

# 可视化结果
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, quadratic_model(x, theta), 'r-', lw=3, label='Fitted Quadratic Curve')
plt.legend()
plt.show()
```
